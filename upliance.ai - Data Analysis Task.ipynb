### Standardize Categorical Columns

```python
user_details['Favorite Meal'] = user_details['Favorite Meal'].str.lower()
cooking_sessions['Meal Type'] = cooking_sessions['Meal Type'].str.lower()
order_details['Meal Type'] = order_details['Meal Type'].str.lower()
```

### Merge Datasets

```python
# Remove leading/trailing spaces
cooking_sessions.columns = cooking_sessions.columns.str.strip()
order_details.columns = order_details.columns.str.strip()
user_details.columns = user_details.columns.str.strip()
```

```python
# Rename columns for consistency
cooking_sessions.rename(columns={'Session ID': 'Session_ID'}, inplace=True)
order_details.rename(columns={'Session ID': 'Session_ID', 'User ID': 'User_ID'}, inplace=True)
user_details.rename(columns={'User ID': 'User_ID'}, inplace=True)
```

```python
merged_data = cooking_sessions.merge(order_details, on='Session_ID', how='inner').merge(user_details, on='User_ID', how='inner')
```

```python
merged_data
```

```python
# Drop the unnecessary columns
merged_data.drop(columns=['Meal Type_y', 'Dish Name_y'], inplace=True)

# Rename the desired columns
merged_data.rename(columns={'Dish Name_x': 'Dish Name', 'Meal Type_x': 'Meal Type'}, inplace=True)

# Verify the changes
print(merged_data.columns)
```

### Data Analysis

**Relationship Between Cooking Sessions and Orders**

```python
merged_data.groupby('User ID')['Duration (mins)'].mean().corr(merged_data.groupby('User ID')['Total Orders'].mean())
```

**Visualizing the relationship**

```python
sns.scatterplot(data=merged_data, x='Duration (mins)', y='Amount (USD)')
plt.show()
```

**Identifying Popular Dishes**

```python
popular_dishes = merged_data['Dish Name'].value_counts()
ax=popular_dishes.plot(kind='bar')
for container in ax.containers:
    ax.bar_label(container)
plt.title('Popular Dishes')
plt.show()
```

**Based on Demographic Factors**

```python
sns.boxplot(data=merged_data, x='Age', y='Amount (USD)')
plt.show()
```

```python
regional_popularity = merged_data.groupby('Location')['Dish Name'].value_counts().unstack().fillna(0)
regional_popularity.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Regional Preferences')
plt.show()
```

**Favorite Meal Distribution**

```python
meal_distribution = user_details['Favorite Meal'].value_counts()

# Plot the pie chart
meal_distribution.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), startangle=90, colors=sns.color_palette('pastel'))
plt.title('Favorite Meal Distribution')
plt.ylabel('')  # Remove the y-label for aesthetics
plt.show()
```

### Heatmap of Spending by Meal Type and Duration

```python
heatmap_data = merged_data.pivot_table(index='Meal Type', columns='Duration (mins)', values='Amount (USD)', aggfunc='mean')
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".1f", cbar_kws={'label': 'Average Amount (USD)'})
plt.title('Heatmap of Spending by Meal Type and Duration')
plt.xlabel('Duration (mins)')
plt.ylabel('Meal Type')
plt.tight_layout()
plt.show()
```
